```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
ds = xr.open_dataset('./grids/residual.nc')

ds['Band1'].plot(vmin=-1000, vmax=1000, cmap='seismic')
```

```python
grid_values = ds['Band1'].data
grid_values[grid_values<-1e9] = np.nan
grid_values = grid_values - np.nanmean(grid_values)

XX,YY = np.meshgrid(ds['x'],ds['y'])

data = np.vstack((XX.flatten(),
                  YY.flatten(),
                  grid_values.flatten())).T
```
