```python
subset = pet[pet.PROPERTY.str.contains('magnetic susceptibility')]

plt.figure(figsize=(22,12))
plt.scatter(subset.SAMPLE_LONGITUDE, subset.SAMPLE_LATITUDE, c=subset.VALUE*10e-5, 
            s=5, vmin=0, vmax=0.1, alpha=0.6)
plt.axis([114,155,-45,-10])
plt.gca().set_aspect('equal')
plt.colorbar()
```

```python
plt.hist(subset.VALUE*10e-5, bins=np.linspace(0,.01,100))
plt.show()
```

```python
import seaborn as sns
from matplotlib import ticker as mticker

subsetl = subset.assign(logVALUE = np.log10(subset.VALUE*10.e-5))
subsetl.replace(-np.inf, np.nan, inplace=True)
subsetl.replace(np.inf, np.nan, inplace=True)
subsetl.dropna(subset=['logVALUE'], inplace=True)
#subsetss = subset[subset.LITHOLOGY.isin(['sand'])]

#grouped_points = subset.groupby(['LITHOLOGY']) 

#plt.figure(figsize=(20,14))
#for group in grouped_points.groups:
#    sus = grouped_points.get_group(group)
#    plt.VALUE
 
fig, ax = plt.subplots(1, 1, figsize=(16,12))
sns.boxplot(x="LITHOLOGY", y="logVALUE", data=subsetl,
            order=['limestone', 'sandstone', 'siltstone', 'mudstone', 'greywacke',
                      'granite', 'basalt', 'gabbro', 
                      'andesite', 'diorite', 'granodiorite', 'rhyolite', 'dacite',
                      'amphibolite', 'gneiss', 'granulite', 'schist', 'serpentinite', 'lherzolite', 'peridotite'],
            whis=[0, 100], width=.6, palette="vlag")
sns.stripplot(x='LITHOLOGY', y='logVALUE', data=subsetl, 
               order=['limestone', 'sandstone', 'siltstone', 'mudstone', 'greywacke',
                      'granite', 'basalt', 'gabbro', 
                      'andesite', 'diorite', 'granodiorite', 'rhyolite', 'dacite',
                      'amphibolite', 'gneiss', 'granulite', 'schist', 'serpentinite', 'lherzolite', 'peridotite'], 
               #inner='point', width=1.5, scale='area')
               alpha=0.2, color='k')

ax.yaxis.set_major_formatter(mticker.StrMethodFormatter("$10^{{{x:.0f}}}$"))
ax.yaxis.set_ticks([np.log10(x) for p in range(-6,1) for x in np.linspace(10**p, 10**(p+1), 10)], minor=True)
plt.grid(axis='y', linestyle=':', which='both')
fig.autofmt_xdate(rotation=45)
#plt.xticks(rotation=30)
plt.show()
```
